```python
import random
import secrets
import os
# PRNG clásico (no seguro para criptografía)
print("PRNG (random):", random.randint(0, 100))
# CSPRNG recomendado para criptografía
print("CSPRNG (secrets):", secrets.randbelow(100))
# RNG basado en el sistema operativo
print("RNG (os.urandom):", int.from_bytes(os.urandom(4), "big"))
```
