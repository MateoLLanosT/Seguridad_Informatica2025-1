1. Implementa una estructura simple de bloque en Python (con campos de datos,
timestamp, hash del bloque anterior y propio).

```python
class Block:
    def __init__(self, index, data, previous_hash):
        
        self.index = index # Índice para identificar la posición del bloque en la cadena
        
        self.timestamp = time.time() # Marca de tiempo de creación del bloque
        
        self.data = data # Información ficticia o transacciones que contiene el bloque
        
        self.previous_hash = previous_hash# Hash del bloque anterior, para mantener la cadena unida

        self.hash = self.calculate_hash() # Calculamos el hash al inicializar el objeto

    def calculate_hash(self):
        block_string = f"{self.index}{self.timestamp}{self.data}{self.previous_hash}"   # Se combinan los campos importantes y se aplica SHA-256
        return hashlib.sha256(block_string.encode()).hexdigest()  # Usamos SHA-256, el estándar en seguridad para blockchains
```

2. Crea una “cadena” añadiendo varios bloques con información ficticia.
3. Verifica cómo se calcula el hash y cómo se encadena la información.
Desafío adicional: Agrega validación de integridad de la cadena (función que revise automáticamente sila cadena es válida).

```python
class Blockchain:
    def __init__(self):
        # La cadena inicia con un bloque génesis
        self.chain = [self.create_genesis_block()] 

    # Primer bloque de la cadena, creado manualmente (previous_hash=0 )
    def create_genesis_block(self):
        return Block(0, "Bloque Génesis", "0")

    def get_latest_block(self):
        return self.chain[-1] # Retorna el último bloque agregado

    def add_block(self, new_data):
        previous_block = self.get_latest_block()
            # Crea un nuevo bloque usando el hash del último bloque
        new_block = Block(len(self.chain), new_data, previous_block.hash)
        self.chain.append(new_block)

    def is_chain_valid(self):
        # Revisa la integridad de la cadena (funcion automatica)
        for i in range(1, len(self.chain)):
            current_block = self.chain[i]
            previous_block = self.chain[i - 1]

            # Verifica que el hash actual no haya cambiado
            if current_block.hash != current_block.calculate_hash():
                return False

            # Verifica que el bloque anterior esté bien enlazado
            if current_block.previous_hash != previous_block.hash:
                return False
        return True
```
